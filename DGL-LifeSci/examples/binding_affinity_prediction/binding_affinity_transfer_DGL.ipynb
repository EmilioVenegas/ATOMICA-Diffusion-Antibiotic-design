```python
# NEW CELL - Load dataset manually (bypassing corrupted files)
from pathlib import Path
from rdkit import Chem
import numpy as np

data_dir = Path.home() / ".dgl" / "pdbbind_v2015" / "v2015"

# Get list of available PDB directories
pdb_dirs = sorted([d for d in data_dir.iterdir() if d.is_dir()])[:50]  # Just first 50

print(f"Loading first {len(pdb_dirs)} PDB entries...\n")

# Create simple dataset structure
class SimplePDBBind:
    def __init__(self, pdb_dirs, max_samples=None):
        self.pdb_dirs = pdb_dirs[:max_samples] if max_samples else pdb_dirs
        self.ligand_mols = []
        self.ligand_files = []
        self.protein_files = []
        self.labels = []
        
        self._load_data()
    
    def _load_data(self):
        """Load ligand and protein files."""
        for i, pdb_dir in enumerate(self.pdb_dirs):
            pdb_name = pdb_dir.name
            
            # Try to find ligand file (prefer mol2, fallback to sdf)
            ligand_file = None
            for ext in ['mol2', 'sdf']:
                f = pdb_dir / f"{pdb_name}_ligand.{ext}"
                if f.exists():
                    ligand_file = f
                    break
            
            # Find protein file
            protein_file = pdb_dir / f"{pdb_name}_pocket.pdb"
            
            if ligand_file and protein_file.exists():
                try:
                    # Load ligand molecule
                    if ligand_file.suffix == '.mol2':
                        mol = Chem.MolFromMol2File(str(ligand_file), sanitize=False, removeHs=False)
                    else:
                        supplier = Chem.SDMolSupplier(str(ligand_file), sanitize=False, removeHs=False)
                        mol = supplier[0] if supplier else None
                    
                    if mol:
                        self.ligand_mols.append(mol)
                        self.ligand_files.append(str(ligand_file))
                        self.protein_files.append(str(protein_file))
                        self.labels.append(0.0)  # Placeholder label
                        
                        if len(self.ligand_mols) == 1:
                            print(f"✓ First sample loaded:")
                            print(f"  PDB: {pdb_name}")
                            print(f"  Ligand atoms: {mol.GetNumAtoms()}")
                            print(f"  SMILES: {Chem.MolToSmiles(mol)[:50]}...")
                except Exception as e:
                    pass
        
        print(f"\n✓ Loaded {len(self.ligand_mols)} valid complexes")
    
    def __len__(self):
        return len(self.ligand_mols)
    
    def __getitem__(self, idx):
        return {
            'ligand_mol': self.ligand_mols[idx],
            'ligand_file': self.ligand_files[idx],
            'protein_file': self.protein_files[idx],
            'label': self.labels[idx]
        }

# Create dataset
dataset = SimplePDBBind(pdb_dirs, max_samples=20)

print(f"\nDataset size: {len(dataset)}")
print(f"\nFirst sample:")
sample = dataset[0]
print(f"  Ligand: {sample['ligand_file']}")
print(f"  Protein: {sample['protein_file']}")
```

```python
# Cell 5: Build model
print("Building PotentialNet model...")
model = PotentialNet(
    f_in=44,
    f_bond=128,
    f_spatial=128,
    f_gather=200,
    n_etypes=9,
    n_bond_conv_steps=3,
    n_spatial_conv_steps=3,
    n_rows_fc=[200, 100],
    dropouts=[0.25, 0.25, 0.25]
)

print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print("Model architecture:")
print(model)
```

```python
# NEW CELL - Create correct dummy graphs with proper edge types
import torch
import dgl

# Set edge type range correctly
num_etypes = 9  # 5 covalent + 4 distance bins

# Create bigraph (Stage 1: covalent bonds)
num_nodes = 50
bigraph = dgl.graph((torch.arange(num_nodes-1), torch.arange(1, num_nodes)))
bigraph.ndata['h'] = torch.randn(num_nodes, 44)

# Set edge types correctly (one-hot encoding, 9 dimensions)
bigraph.edata['e'] = torch.zeros(num_nodes-1, num_etypes)
bigraph.edata['e'][:, 0] = 1.0  # Type 0 (single bond)

print(f"Bigraph edge types shape: {bigraph.edata['e'].shape}")
print(f"Max edge type index: {bigraph.edata['e'].argmax(dim=1).max().item()}\n")

# Create knn_graph (Stage 2: KNN + covalent)
k = 10
src = torch.arange(num_nodes).repeat(k)
dst = torch.arange(k).repeat(num_nodes).flatten()
knn_graph = dgl.graph((src, dst))

# Set edge types for KNN graph (also 9 dimensions)
knn_graph.edata['e'] = torch.zeros(len(src), num_etypes)
knn_graph.edata['e'][:, 0] = 1.0  # All type 0

print(f"KNN graph edge types shape: {knn_graph.edata['e'].shape}")
print(f"Max edge type index: {knn_graph.edata['e'].argmax(dim=1).max().item()}")

# Now try forward pass
print("\nTesting forward pass...")
try:
    with torch.no_grad():
        prediction = model(bigraph, knn_graph)
    print(f"✓ Success! Prediction: {prediction.item():.4f}")
except Exception as e:
    print(f"✗ Error: {e}")
    print("\nCheck edge type indices are in range [0, 9)")
```
